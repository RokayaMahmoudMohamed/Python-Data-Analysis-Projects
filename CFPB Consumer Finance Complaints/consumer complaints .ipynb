**ignore the column we will not need in analysis**

```python
#0 1 3 7 8 12 13 14 15 16 17
dt=pd.read_csv('consumer_complaints.csv',
               usecols=[0,1,3,7,8,12,13,14,15,16,17],
               dtype=np.str)
dt.head()
```

**delete null values of state**

```python
np.sum(dt['state'].isnull())
```

```python
consumer_complaints=dt[dt['state'].notna()]
```

```python
consumer_complaints.info()
```

```python
consumer_complaints
```

 **check for valid sending and receiving date  the consumer complaints**
 

**note that there is value of received data before send date and that is not correct**

```python
np.sum((consumer_complaints['date_received']>consumer_complaints['date_sent_to_company'])==True)
```

```python
cons_comp=consumer_complaints[consumer_complaints['date_received']<=consumer_complaints['date_sent_to_company']]
cons_comp
```

```python
np.sum((cons_comp['date_received']>cons_comp['date_sent_to_company'])==True)  ##so that's right 
```

```python
#take data to csv file
cons_comp.to_csv('consumer_complaints_cleaned.csv')
```
